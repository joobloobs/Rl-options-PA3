```python
from util import *
from Agent import SMDPAgent, TaxiAgent, IntraOptAgent
import gymnasium as gym
```

```python
def experience(agent: TaxiAgent, n_iter=10000, offset=0):
    env = gym.make('Taxi-v3')
    rewards = agent.train(n_iter)[offset:]
    plt.xlabel("episode number")
    plt.ylabel("reward")
    plt.title("Reward vs episode")
    plt.plot(range(offset, n_iter), rewards)
    plt.show()
    env.close()
```

## SMDP q-learning with options to go to red, green, blue, yellow

```python
agent = SMDPAgent()
experience(agent, offset=100)
plot_Q(agent.options)
```

## Intra q-learning

```python
agent = IntraOptAgent()
experience(agent, offset=100)
plot_Q(agent.options)
```

## SMDP q-learning with get and put

The 2 options are get and put :
- get finishes when the passenger is in the taxi
- put finishes when passenger is no longer in taxi

```python
agent = SMDPAgent(get_put_options=True)
experience(agent, offset=100)
plot_Q(agent.options)
```

## Intra q-learning with get and put

```python
agent = IntraOptAgent(get_put_options=True)
experience(agent, offset=100)
plot_Q(agent.options)
```

## Inferences
The second set of options works better than the given options as they go straight to the goal : picking up the passenger and putting it to the destination.
Intra q-learning performs better in the sense that it allows to learn faster / to stabilize faster which is logical as it updates the q_values more often
The policies learned for the options doesn't change between SMDP q-learning and Intra q-learning.
